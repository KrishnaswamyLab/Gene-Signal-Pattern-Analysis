```python
import scanpy
import pandas as pd
import numpy as np
import phate

import sys
sys.path.append('..')
from run.run_gspa import calculate_wavelet_dictionary
from run.run_ae_default_config import run_ae
from utils import *
import seaborn as sns
```

```python
data_int = scanpy.read_h5ad('data/int.h5ad')
data_gc = scanpy.read_h5ad('data/GC.h5ad')
hits = pd.read_csv('data/tf_hits.csv', index_col=0)
guide_hit_signal = hits.columns[np.where(hits.values > 0)[1]]
```

```python
for guide in hits.columns:
    print (guide)
    data = data_int[guide_hit_signal == guide].to_df()
    # get graph of cells for hit
    phate_op = phate.PHATE(random_state=42, n_jobs=-1, knn=30, verbose=False)
    NEG_phate = phate_op.fit_transform(data)
    G = phate_op.graph.to_pygsp()
    
    # get cell dictionary
    cell_dictionary, wavelet_sizes = calculate_wavelet_dictionary(G, use_reduced=False)

    # get uniform signal
    uniform_signal = np.ones((1, G.N))
    uniform_signal = uniform_signal / np.linalg.norm(uniform_signal, axis=1).reshape(-1,1)


    # normalize signals and get those with norm > 0
    data = data.iloc[:, np.linalg.norm(data, axis=0) > 0] / np.linalg.norm(data.iloc[:, np.linalg.norm(data, axis=0) > 0], axis=0)

    # get signal embedding
    signals_projected = project(data.T, cell_dictionary)
    signals_reduced = svd(signals_projected)
    #signals_ae = run_ae(signals_reduced)
    
    # get uniform representation
    uniform_projected = project(uniform_signal, cell_dictionary)
    
    # save
    results = {}
    results['signal_embedding'] = signals_reduced[:, :5]
    results['localization_score'] = calculate_distance_to_signal(uniform_projected, signals_projected)
    np.savez(f'results/{guide}_GSPA.npz', signal_embedding=results['signal_embedding'],
             localization_score=results['localization_score'], genes=data.columns.values)
```
